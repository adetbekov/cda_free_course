# Различные виды распределений вероятностей: нормальное, равномерное, экспоненциальное и т.д.

## Импорт необходимых библиотек

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Создание набора данных


```python
# Генерация случайных данных, соответствующих нормальному распределению
# среднее значение и стандартное отклонение
mu, sigma = 0, 0.1
data_normal = np.random.normal(mu, sigma, 1000)

# Генерация случайных данных, соответствующих равномерному распределению
data_uniform = np.random.uniform(-1, 1, 1000)

# Генерация случайных данных, соответствующих экспоненциальному распределению
data_exponential = np.random.exponential(1, 1000)
```

```python
plt.hist(data_exponential, bins=10)
```

## Построение гистограмм для каждого распределения

```python
plt.hist(data_normal, bins=50, alpha=0.5, label='Нормальное распределение')
plt.hist(data_uniform, bins=50, alpha=0.5, label='Равномерное распределение')
plt.hist(data_exponential, bins=50, alpha=0.5, label='Экспоненциальное распределение')
plt.legend()
plt.show()
```
